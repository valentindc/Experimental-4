```python
import csv
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import scipy.constants as cte
import scipy.odr.odrpack as odr


def u(B, x):
    return B[0]*x + B[1]


linear = odr.Model(u)


def fit (f, x, y):
    popt, pcov = curve_fit(f, x, y)
    o = popt
    p = pcov
    return o, p


voltage0 = []
current0 = []
freq = []


x_e = np.linspace( 0.25, 15, 50, endpoint = False)
        

ierr =[]
verr = []
with open('1260hz.csv', 'r') as file:
    reader = csv.reader(file)
    freq0 = 12.6132e14 #[freq] = hz
    freq.append(freq0)
    for row in reader:
        voltage0.append(float(row[2])) #[V] = volts
        current0.append(float(row[3])) #[i]= pA = e-12A 
        verr.append(0.002) #el error de exactitud del volt√≠metro
        ierr.append(0.02)  #again, error de exactitud


mydata0 = odr.RealData( current0, voltage0, sx= ierr, sy= verr)
myodr0 = odr.ODR(mydata0, linear, beta0=[-1, 2])
output0 = myodr0.run()
#output0.pprint()
plt.errorbar(current0, voltage0, xerr= ierr, yerr= verr, fmt="go")
plt.plot(x_e, u( output0.beta, x_e), "g-")



voltage11 = []
current11 = []
with open('1200hz.csv', 'r') as file:
    reader = csv.reader(file)
    freq11 = 11.9573e14 #[freq] = hz
    freq.append(freq11)
    #metal = row[1]
    for row in reader:
        voltage11.append(float(row[2])) #[V] = volts
        current11.append(float(row[3])) #[i]= pA = e-12A 

mydata11 = odr.RealData( current11, voltage11, sx= ierr, sy= verr)
myodr11 = odr.ODR(mydata11, linear, beta0=[-1, 2])
output11 = myodr11.run()
plt.errorbar(current11, voltage11, xerr= ierr, yerr=verr, fmt="mo")
plt.plot(x_e, u(output11.beta, x_e), "m-")



voltage1 = []
current1 = []
with open('1140hz.csv', 'r') as file:
    reader = csv.reader(file)
    freq1 = 11.4112e14 #[freq] = hz
    freq.append(freq1)
    #metal = row[1]
    for row in reader:
        voltage1.append(float(row[2])) #[V] = volts
        current1.append(float(row[3])) #[i]= pA = e-12A 

mydata1 = odr.RealData( current1, voltage1, sx= ierr, sy= verr)
myodr1 = odr.ODR(mydata1, linear, beta0=[-1, 2])
output1 = myodr1.run()
plt.errorbar(current1, voltage1, xerr= ierr, yerr=verr, fmt="ro")
plt.plot(x_e, u(output1.beta, x_e), "r-")




voltage21 = []
current21 = []
with open('1100hz.csv', 'r') as file:
    reader = csv.reader(file)
    freq21 = 10.9625e14 #[freq] = hz
    freq.append(freq21)
    #metal = row[1]
    for row in reader:
        voltage21.append(float(row[2])) #[V] = volts
        current21.append(float(row[3])) #[i]= pA = e-12A 
        
mydata21 = odr.RealData( current21, voltage21, sx= ierr, sy= verr)
myodr21 = odr.ODR(mydata21, linear, beta0=[-1, 2])
output21 = myodr21.run()
plt.errorbar(current21, voltage21, xerr= ierr, yerr=verr, fmt="yo")
plt.plot(x_e, u( output21.beta, x_e), "y-")



voltage2 = []
current2 = []
with open('1050hz.csv', 'r') as file:
    reader = csv.reader(file)
    freq2 = 10.5165e14 #[freq] = hz
    freq.append(freq2)
    #metal = row[1]
    for row in reader:
        voltage2.append(float(row[2])) #[V] = volts
        current2.append(float(row[3])) #[i]= pA = e-12A 
        
mydata2 = odr.RealData( current2, voltage2, sx= ierr, sy= verr)
myodr2 = odr.ODR(mydata2, linear, beta0=[-1, 2])
output2 = myodr2.run()
plt.errorbar(current2, voltage2, xerr= ierr, yerr=verr, fmt="bo")
plt.plot(x_e, u(output2.beta, x_e), "b-")



voltage31 = []
current31 = []
with open('990hz.csv', 'r') as file:
    reader = csv.reader(file)
    freq31 = 9.8957e14 #[freq] = hz
    freq.append(freq31)
    #metal = row[1]
    for row in reader:
        voltage31.append(float(row[2])) #[V] = volts
        current31.append(float(row[3])) #[i]= pA = e-12A 
        
mydata31 = odr.RealData( current31, voltage31, sx= ierr, sy= verr)
myodr31 = odr.ODR(mydata31, linear, beta0=[-1, 2])
output31 = myodr31.run()
plt.errorbar(current31, voltage31, xerr= ierr, yerr=verr, fmt="ko")
plt.plot(x_e, u( output31.beta, x_e), "k-")




voltage3 = []
current3 = []
with open('961hz.csv', 'r') as file:
    reader = csv.reader(file)
    freq3 = 9.6120e14 #[freq] = hz
    freq.append(freq3)
    #metal = row[1]
    for row in reader:
        voltage3.append(float(row[2])) #[V] = volts
        current3.append(float(row[3])) #[i]= pA = e-12A 
        
mydata3 = odr.RealData( current3, voltage3, sx= ierr, sy= verr)
myodr3 = odr.ODR(mydata3, linear, beta0=[-1, 2])
output3 = myodr3.run()
plt.errorbar(current3, voltage3, xerr= ierr, yerr=verr, fmt="co")
plt.plot(x_e, u( output3.beta, x_e), "c-")


plt.savefig("V(i).jpg", dpi= 250)
```
